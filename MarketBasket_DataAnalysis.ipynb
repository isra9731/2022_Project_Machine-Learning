##Market Basket Analysis For Online Retail Dataset

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Reading Data from Web
MyRetailData=pd.read_csv('/content/OnlineRetail.csv',encoding= 'unicode_escape')
MyRetailData.head()
```

```python
MyRetailData.shape
```

```python
MyRetailData.columns
```

###**Data Preparation**

```python
#Data Cleaning
MyRetailData['Description'] = MyRetailData['Description'].str.strip() #removes extra spaces
```

```python
#Removing duplicate invoiceno
MyRetailData.dropna(axis=0, subset=['InvoiceNo'], inplace=True) #inplace=true tells to make changes in the same data rather then creating a different copy
```

```python
#Converting my invoice as a string
MyRetailData['InvoiceNo']=MyRetailData['InvoiceNo'].astype('str')
```

```python
#remove credit transaction
MyRetailData=MyRetailData[~MyRetailData['InvoiceNo'].str.contains('C')]
```

```python
MyRetailData
```

```python
MyRetailData['Country'].value_counts()
```

```python
basket=(MyRetailData[MyRetailData['Country']=="Germany"]
        .groupby(['InvoiceNo','Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
#converts numbers < 0 is 0 and numbers >1 as 1
def my_encode_units(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1

basket_sets=basket.applymap(my_encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

###**Training Model**

```python
#Generate Frequent Itemsets
frequent_itemsets=apriori(basket_sets,min_support=0.07,use_colnames=True) #finds frequent items from the dataset
```

```python
#Building Association Rules Using Frequent Itemset
rules_frequent_itemsets=association_rules(frequent_itemsets, metric="lift", min_threshold = 1) #metric : measure of importance of rule
```

```python
rules_frequent_itemsets.head(100)
```

###Making Recommendation
